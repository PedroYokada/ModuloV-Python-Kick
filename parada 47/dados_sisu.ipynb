```python
import pandas as pd
import matplotlib.pyplot as plt

# linha que carrega o arquivo csv
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep = '|' , encoding= 'latin-1')

# dois metodos
# leitura das primeiras linhas do arquivo
print(df.head())
print(df.columns)
# pegar o nome das colunas da tabela


campus_por_estado = df ['UF_CAMPUS'].value_counts()
print(campus_por_estado)

cursos = df['NOME_CURSO'].value_counts()
print(cursos)

campus_por_estado.plot(kind='bar')
plt.xlabel('UF_CAMPUS')
plt.title('Distribuição de Campus por Estado')
plt.show()

cursos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos')
plt.show()

top_10_cursos = cursos.sort_values(ascending=False).head(10)
top_10_cursos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos - TOP 10')
plt.show()

cota = df ['TP_COTA'].value_counts()
print(cota)












```
